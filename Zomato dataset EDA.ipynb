```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
#CHECK datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
counrty_names= final_df.Country.value_counts().index
```

```python
Country_val=final_df.Country.value_counts().values
```

```python
# pie chart top three
plt.pie(Country_val[:3],labels=counrty_names[:3],autopct='%1.2f%%');

#max no of transactions happen in india
```

Observation : Zomato maximum records or transcations are from India.
After than US and then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observtions
1. When rating is between 4.5 - 5 ---> Excellent
2. When rating is between 4.0 - 4.4 ---> Very good
3. When rating is between 3.5 - 3.9 ---> Good
4. When rating is between 3.0 - 3.4 ---> Average
5. When rating is between 2.5 - 2.9 ---> Average
6. When rating is between 1.8 - 2.4 ---> Poor
7. 2148 either have given 0 rating or have not given any rating 

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color',data=ratings,palette=['grey','red','orange','yellow','green','green'])
```

## Observation
1. Not rated count is very high
2. Max number of ratings and between 2.5 - 3.4 # if the values were missing we could have used these inplace

```python
# countplot
sns.countplot(x='Rating color',data=ratings,palette=['grey','red','orange','yellow','green','green']); # these are frequrncy of colour not no of ratimgs
```
